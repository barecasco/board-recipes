```python
management = [
    {"mpa": "A", "manager": "Jimy", "luasan_mpa": 100},
    {"mpa": "B", "manager": "Asri", "luasan_mpa": 100},
    {"mpa": "C", "manager": "Diah", "luasan_mpa": 150}
]

management_df = pd.DataFrame(management)
management_df
```

```python
ecosystem = [
    {"mpa" : "A", "ecosystem" : "karang", "luasan_ecosystem": 30},
    {"mpa" : "B", "ecosystem" : "karang", "luasan_ecosystem": 15},
    {"mpa" : "C", "ecosystem" : "mangrove", "luasan_ecosystem": 20},
    {"mpa" : "A", "ecosystem" : "lamun", "luasan_ecosystem": 40},
    {"mpa" : "C", "ecosystem" : "karang", "luasan_ecosystem": 10},
]

ecosystem_df = pd.DataFrame(ecosystem)
ecosystem_df
```

```python
rowdatas = {}

for i, row in management_df.iterrows():
    mpa        = row["mpa"]
    manager    = row["manager"]
    luasan_mpa = row["luasan_mpa"]
    rowdatas[mpa, manager, luasan_mpa] = []
    
    for j, ecorow in ecosystem_df.iterrows():
        ecompa           = ecorow["mpa"]
        ecosystem        = ecorow["ecosystem"]
        luasan_ecosystem = ecorow["luasan_ecosystem"]
        if (mpa == ecompa) :
            rowdatas[mpa, manager, luasan_mpa].append([ecompa, ecosystem, luasan_ecosystem])
```

```python
rowdatas
```

```python
rows = []

for key, children in rowdatas.items():
    for child in children:
        row_item = {
            "mpa"        : key[0],
            "manager"    : key[1],
            "luasan_mpa" : key[2],
            "ecosystem"  : child[1],
            "luasan_ecosystem" : child[2]
        }
        rows.append(row_item)

final_df = pd.DataFrame(rows)

```
